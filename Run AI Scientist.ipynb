```python
# Run LLM reviewer
import openai
from ai_scientist.perform_review import load_paper, perform_review

client = openai.OpenAI()
model = "deepseek_chat"

# Load paper from PDF file (raw text)
paper_txt = load_paper("result/deepfake/20241211_140714_Parameterise analytic filterbanks in RawNet2/Parameterise analytic filterbanks in RawNet2.pdf")

# Get the review dictionary
review = perform_review(
    paper_txt,
    model,
    client,
    num_reflections=5,
    num_fs_examples=1,
    num_reviews_ensemble=5,
    temperature=0.1,
)

# Inspect review results
review["Overall"]    # Overall score (1-10)
review["Decision"]   # 'Accept' or 'Reject'
review["Weaknesses"] # List of weaknesses (strings)

# Load paper from PDF file (raw text)
paper_txt = load_paper("result/deepfake/20241211_132310_channel_wise_attention_in_rawnet2/channel_wise_attention_in_rawnet2.pdf")

# Get the review dictionary
review = perform_review(
    paper_txt,
    model,
    client,
    num_reflections=5,
    num_fs_examples=1,
    num_reviews_ensemble=5,
    temperature=0.1,
)

# Inspect review results
review["Overall"]    # Overall score (1-10)
review["Decision"]   # 'Accept' or 'Reject'
review["Weaknesses"] # List of weaknesses (strings)
```
